```python
import random
import csv
from datetime import datetime, timedelta

# Listas de dados para geração dos registros
nomes_produtos = ["Panela Antiaderente", "Creme Facial", "Ração para Cães", "Smartphone", "Boneca", "Chaleira Elétrica", "Shampoo", "Mouse Gamer", "Quebra-Cabeça", "Fone de Ouvido"]
categorias = ["Casa e Cozinha", "Saúde e Beleza", "Pets", "Eletrônicos", "Brinquedos"]
setores = ["Oferta da Semana", "Lançamentos", "Promoções", "Nenhum"]
tipos_pagamento = ["A vista", "Crédito"]
formas_pagamento = ["Visa", "Mastercard", "Elo", "American Express", "Boleto", "Pix"]

# Função para gerar um registro aleatório
def gerar_registro():
    nome_produto = random.choice(nomes_produtos)
    categoria = random.choice(categorias)
    setor = random.choice(setores)
    data_compra = (datetime.now() - timedelta(days=random.randint(1, 365))).strftime("%d/%m/%Y")
    valor_compra = round(random.uniform(10, 500), 2)
    quantidade_produto = random.randint(1, 10)
    tipo_pagamento = random.choice(tipos_pagamento)
    forma_pagamento = random.choice(formas_pagamento)
    return [nome_produto, categoria, setor, data_compra, valor_compra, quantidade_produto, tipo_pagamento, forma_pagamento]

# Gera os registros
registros = []
for _ in range(5000):
    registros.append(gerar_registro())

# Escreve os registros em um arquivo CSV
with open('registros_dropshipping5.csv', mode='w', newline='', encoding='utf-8') as arquivo_csv:
    writer = csv.writer(arquivo_csv, delimiter=';')
    writer.writerow(["Nome do produto", "Categoria", "Setor", "Data da compra", "Valor da compra", "Quantidade de produto", "Tipo de pagamento", "Forma de pagamento"])
    writer.writerows(registros)

print("Registros gerados e salvos em registros_dropshipping.csv.")
```

```python
import random
import json
import requests

def generate_fake_clients(num_clients):
    url = 'https://servicodados.ibge.gov.br/api/v1/localidades/estados'
    response = requests.get(url)
    states = json.loads(response.text)

    clients = []

    for _ in range(num_clients):
        state = random.choice(states)
        state_name = state['nome']
        state_initials = state['sigla']

        url = f'https://servicodados.ibge.gov.br/api/v1/localidades/estados/{state_initials}/municipios'
        response = requests.get(url)
        cities = json.loads(response.text)
        city = random.choice(cities)
        city_name = city['nome']

        phone_number = generate_phone_number(state_initials)

        client = {
            'city': city_name,
            'state': state_initials,
            'phone': phone_number
        }

        clients.append(client)

    return clients

def generate_phone_number(state_initials):
    ddds = {
        'AC': '68', 'AL': '82', 'AP': '96', 'AM': '92,97', 'BA': '71,73,74,75,77',
        'CE': '85,88', 'DF': '61', 'ES': '27,28', 'GO': '62,64', 'MA': '98,99',
        'MT': '65,66', 'MS': '67', 'MG': '31,32,33,34,35,37,38', 'PA': '91,93,94',
        'PB': '83', 'PR': '41,42,43,44,45,46', 'PE': '81,87', 'PI': '86,89',
        'RJ': '21,22,24', 'RN': '84', 'RS': '51,53,54,55', 'RO': '69',
        'RR': '95', 'SC': '47,48,49', 'SP': '11,12,13,14,15,16,17,18,19',
        'SE': '79', 'TO': '63'
    }

    ddd = random.choice(ddds[state_initials].split(','))
    number = ''.join(random.choices('0123456789', k=8))

    return f'({ddd}) 9{number[:4]}-{number[4:]}'

# Gerar 500 registros de clientes falsos
clients = generate_fake_clients(500)

# Exibir os dados dos clientes gerados
for client in clients:
    print(client)
```
